# Analysis of Titanic_dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("titanic_Data_Train1.csv")

# Display the first few rows of the dataframe
print("First few rows of the dataset:")
print(df.head())
```

```python

# Drop the irrelevant columns
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns', inplace=True)

# Display the first few rows of the dataframe after dropping columns
print("\nDataset after dropping irrelevant columns:")
print(df.head())
```

```python

# Check for missing values
print("\nMissing values in the dataset:")
print(df.isnull().sum())

# Handle missing values (for simplicity, fill missing Age values with the median and drop rows with missing Fare values)
df['Age'].fillna(df['Age'].median(), inplace=True)
df.dropna(subset=['Fare'], inplace=True)

# Verify that there are no more missing values
print("\nMissing values after handling:")
print(df.isnull().sum())

# Summary statistics of the dataset
print("\nSummary statistics:")
print(df.describe())

```

```python
# Exploratory Data Analysis
# Plot the distribution of numerical features
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde=True)
plt.title('Distribution of Age')

plt.subplot(1, 2, 2)
sns.histplot(df['Fare'], kde=True)
plt.title('Distribution of Fare')

plt.tight_layout()
plt.show()

# Plot the survival rate based on Sex
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival Rate by Sex')
plt.show()

# Convert categorical variable 'Sex' into dummy/indicator variables
df = pd.get_dummies(df, columns=['Sex'], drop_first=True)

# Display the first few rows of the dataframe after encoding
print("\nDataset after encoding categorical variables:")
print(df.head())
```
