# Popularity Based Recommendation System

```python
ratings_and_books = ratings.merge(books,on="ISBN")
```

```python
ratings_and_books.shape
```

```python
ratings_and_books.head(5)
```

```python
ratings_and_books[ratings_and_books['ISBN'] == '0001848445']
```

```python
num_rating = ratings_and_books.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating.rename(columns={'Book-Rating' : 'Num-Rating'},inplace=True)
num_rating
```

```python
avg_rating = ratings_and_books.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating' : 'Avg-Rating'},inplace=True)
avg_rating
```

```python
popular = num_rating.merge(avg_rating,on='Book-Title')
```

```python
popular
```

```python
popular.sort_values(by=['Num-Rating'],ascending=False)
```

```python
popular = popular[popular['Num-Rating']>=250].sort_values(by=['Avg-Rating'],ascending=False).head(50)
```
